```python
import networkx as nx
import matplotlib.pyplot as plt
print(nx.__version__)
```

```python
# Graph nodes
G = nx.Graph()
nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]
G.add_nodes_from(nodes)

#Graph edges
edges = [
    ("Andre", "Carol"), ("Andre", "Diane"), ("Andre", "Fernando"),
    ("Beverly", "Diane"), ("Beverly", "Ed"),
    ("Carol", "Diane"), ("Carol", "Fernando"),
    ("Diane", "Fernando"), ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Heather"),
    ("Heather", "Ike"), ("Ike", "Jane")
]
G.add_edges_from(edges)

print(G.nodes(), G.edges())
```

```python
# Draw the graph
plt.figure(figsize=(8,6))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color="skyblue", edge_color="purple", node_size=3000, font_size=12, font_weight="bold")
plt.title("Krackhardt Kite Graph")
plt.show()
```
